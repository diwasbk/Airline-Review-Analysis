## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Cleaning

```python
data = pd.read_csv("AirlineReviews.csv")
# Droping the rows with any missing values
cleaned_data = data.dropna()
cleaned_data.to_csv('cleaned_sample_data.csv', index=False)
```

## Loading Data into DataFrame

```python
df = pd.read_csv("cleaned_sample_data.csv")
```

## Line Graph

```python
filtered_data = df[df['OverallScore'] > 5]  # Example filter: OverallScore greater than 5
data = filtered_data[['DateFlown', 'OverallScore']].head(50)
plt.figure(figsize=(12, 6))
plt.plot(data['DateFlown'], data['OverallScore'], marker='o', linestyle='-', label='Overall Score')
plt.title('Overall Score Over Time')
plt.xlabel('Date Flown')
plt.ylabel('Overall Score')
plt.xticks([])  # Hides the x-axis labels
plt.tick_params(axis='x', bottom=False, top=False, labelbottom=False, labeltop=False)
plt.tight_layout()
plt.show()
```

This line graph visualizes trends in overall score over date flown tracking changes in 
customer in customer satisfaction over time. 

## Bar Graph

```python
data = df.head(50)
sample_data = data.head(50)
plt.figure(figsize=(12, 6))
plt.bar(sample_data['AirlineName'], sample_data['OverallScore'])
plt.title('Overall Score for Sample Airlines')
plt.xlabel('Airline Name')
plt.ylabel('Overall Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This bar graph compares overall scores across different airline names, clearly showing 
performance differences among airlines. 

## Histogram

```python
plt.figure(figsize=(12, 6))
plt.hist(df['OverallScore'].dropna(), bins=20, edgecolor='black')
plt.title('Distribution of Overall Scores')
plt.xlabel('Overall Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

This histogram shows the distribution of overall scores according to the frequency and 
range of satisfaction levels. 

## Box-Whisker Plot

```python
plt.figure(figsize=(12, 6))
plt.boxplot(df['OverallScore'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='red', color='black'))
plt.title('Box-Whisker Plot of Overall Scores')
plt.xlabel('Overall Score')
plt.tight_layout()
plt.show()
```

This box-whisker plot summarizes the overall score distribution highlighting the median, 
quartiles, and outliers. 

## Pie Chart

```python
df = pd.read_csv("cleaned_sample_data.csv")
top_countries = df['OriginCountry'].value_counts().head(5)
top_airlines = df['AirlineName'].value_counts().head(5)
plt.figure(figsize=(16, 16))
plt.subplot(1, 2, 1)
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', textprops={'fontsize': 14})
plt.title('Top 5 Origin Countries', fontweight='bold', fontsize=14)
plt.subplot(1, 2, 2)
plt.pie(top_airlines, labels=top_airlines.index, autopct='%1.1f%%', textprops={'fontsize': 14})
plt.title('Top 5 Airlines', fontweight='bold', fontsize=14)
plt.tight_layout()
plt.show()
```

This pie charts shows the distribution of origin country and airline, illustrating the 
proportion of each category within the dataset.
